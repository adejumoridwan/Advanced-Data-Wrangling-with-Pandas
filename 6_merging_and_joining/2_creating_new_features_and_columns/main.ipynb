```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns

# Load Iris dataset
data = sns.load_dataset("iris")
# Displaying the first few rows of the dataset
print("Original Dataset:")
print(data.head())
```

```python


# Creating a new column 'sepal_ratio' by dividing sepal length by sepal width
data['sepal_ratio'] = data['sepal_length'] / data['sepal_width']
```

```python
data.head()
```

```python
# Defining a custom function to categorize petal length
def categorize_petal_length(length):
    if length < 3:
        return 'Short'
    elif length >= 3 and length < 5:
        return 'Medium'
    else:
        return 'Long'
```

```python

# Applying the custom function to create a new column 'petal_length_category'
data['petal_length_category'] = data['petal_length'].apply(categorize_petal_length)
```

```python
data.head()
```

```python
# Using lambda function to create a new column 'sepal_area'
data['sepal_area'] = data.apply(lambda row: row['sepal_length'] * row['sepal_width'], axis=1)
```
